```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residenciais_limpo.csv', sep = ';')
```

### Criar uma nova coluna que tenha o valor bruto das despesas mensais
#### Valor + Condomínio + IPTU

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

## Valor por metro²

```python
dados['Valor M²'] = dados['Valor']/dados['Area']
```

```python
dados
```

## <font color ="FF8000">.round</font> irá arrendondar as casas decimáis acrescentadas

```python
dados['Valor M²'] = dados['Valor M²'].round(2)
```

```python
dados
```

## Valor Bruto por M²

```python
#para o .round funcionar precisa colocar a operação em parenteses, por ter mais de 1
dados['Valor Bruto M²'] = (dados['Valor Bruto']/dados['Area']).round(2)
```

```python
dados
```

# Criar variável que contenha casa + apartamento

## <font color ="FF8000"> .apply </font> permite que aplique uma função em cada linha/registro do dataframe

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
#a linha de lambda x - vai ser 'Casa' se 'x' estiver dentro da lista casa. Se não vai ser Apartamento.
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo as colunas

#### Como a coluna de IPTU não possui sempre valores válidos, as colunas de Valor Bruto e Valor Bruto M² precisam ser eliminadas

Cria um banco de dados auxiliar

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor M²', 'Valor Bruto', 'Valor Bruto M²']])
```

```python
dados_aux.head(10)
```

## <font color ="FF8000">del</font>

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

## <font color ="FF8000"> pop </font>

```python
dados_aux.pop('Valor Bruto M²')
```

## <font color ="FF8000"> drop </font>

### permite excluir multiplas variáveis/colunas ou linhas
### para coluna axis = 1
### para linha axis = 0

Esse método será usado direto na base de dados original que queremos tratar, e não na auxiliar

```python
dados.drop(['Valor Bruto', 'Valor Bruto M²'], axis = 1, inplace = True)
```

```python
dados.head(10)
```

```python
dados.to_csv('dados/aluguel_residenciais_m2.csv', sep = ';', index = False)
```
