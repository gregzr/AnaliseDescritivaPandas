```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residenciais_m2.csv', sep = ';')
```

```python
dados.head(10)
```

#### Média dos valores

```python
dados['Valor'].mean()
```

#### Seleciona os bairros que quero trabalhar

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
```

#### Vai colocar todos os bairros listados acima como verdadeiros

```python
selecao = dados['Bairro'].isin(bairros)
```

```python
selecao
```

```python
dados = dados[selecao]
```

```python
dados
```

```python
dados['Bairro'].drop_duplicates()
```

### Criar o primeiro grupo com as variáveis do Bairro.

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

#### Cada grupo é um bairro. Como um dicionário.

```python
grupo_bairro.groups
```

```python
for bairro, dados in grupo_bairro:
    print(bairro)
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, dados['Valor'].mean()))
```

# <font color = 'ff8000'> aggregate </font>

### Para selecionar qual estatística deve ser usada.

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum'])
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns={'min': 'Mínimo', 'max':'Máximo', 'sum':'Soma'})
```

Plotando o desvio padrão (std) em gráfico

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```

# Exercício

```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto', sort = False)
produtos.describe().round(2)
```

```python
estatisticas = ['mean', 'std', 'min', 'max']
nomes = {'mean': 'Média', 'std': 'Desvio Padrão', 
    'min': 'Mínimo', 'max': 'Máximo'}
produtos['Preço'].aggregate(estatisticas).rename(columns = nomes).round(2)
```
